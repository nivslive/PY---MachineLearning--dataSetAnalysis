O código abaixo carregará o conjunto de dados "Iris", que é um exemplo com informações sobre diferentes tipos de flores iris e suas medidas. Isso pode ser usado para análises de dados e treinamento de modelos de aprendizado de máquina.

```python
from sklearn.datasets import load_iris
return_load_iris = load_iris()
print(return_load_iris)
```

A função putType recebe um valor numérico que representa uma classe de flor íris e retorna o nome da espécie correspondente ('setosa', 'versicolor' ou 'virginica'). Isso ajuda a traduzir os valores numéricos das classes em nomes mais compreensíveis.

```python
def putType(target):
    if(target == 0):
        return 'setosa'
    elif(target == 1):
        return 'versicolor'
    elif(target == 2):
        return 'virginica'
```

```python
# Criar um DataFrame 'iris' com os dados do conjunto de dados Iris e definir colunas
iris = pd.DataFrame(return_load_iris.data)
iris.columns = ['comprimento_sépala', 'largura_sépala', 'comprimento_pétala', 'largura_pétala']

# Adicionar a coluna 'target' contendo os rótulos numéricos das classes
iris['target'] = return_load_iris.target

# Aplicar a função putType para criar a coluna 'target_name' com os nomes das espécies
iris['target_name'] = iris['target'].apply(putType)

# Exibir a tabela resultante das filtragens
display(iris)

# Importar a biblioteca seaborn e configurar o estilo da visualização
import seaborn as sns
sns.set(rc={'figure.figsize': (15, 12)})
sns.set_theme(style="white")

# Criar um boxplot para análise exploratória dos dados
ax = sns.boxplot(data=iris)

# Verificar os registros que são outliers na largura da sépala
outliers = iris[(iris['largura_sépala'] < 2.2) | (iris['largura_sépala'] > 4)]
```

Agora após a análise exploratória dos nossos dados, iremos fazer os treinos de modelos para a validação do mesmo.

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(return_load_iris.data,return_load_iris.target,random_state=1)
```

```python
treino = pd.DataFrame(x_treino)
treino.columns = ['comprimento_sépala',
                  'largura_sépala',
                  'comprimento_pétala',
                  'largura_pétala'
                ]


treino['target'] = y_treino
display(treino)


sns.scatterplot(x='largura_pétala',y='comprimento_pétala',data=treino, hue='target',palette=['green','orange','blue'])
plt.show()
```
